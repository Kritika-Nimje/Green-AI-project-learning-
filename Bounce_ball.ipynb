```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Bouncing Ball")

# Colors
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# Clock for controlling the frame rate
clock = pygame.time.Clock()

# Ball properties
ball_radius = 20
ball_x = WIDTH // 2
ball_y = ball_radius
ball_velocity_y = 0

gravity = 0.5
bounce_damping = 0.8  # Energy lost on each bounce

# Ground level
ground_y = HEIGHT - ball_radius

running = True
while running:
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Physics update
    ball_velocity_y += gravity
    ball_y += ball_velocity_y

    # Bounce when hitting the ground
    if ball_y >= ground_y:
        ball_y = ground_y
        ball_velocity_y = -ball_velocity_y * bounce_damping

        # Stop bouncing if velocity is very low
        if abs(ball_velocity_y) < 1:
            ball_velocity_y = 0

    # Drawing
    screen.fill(WHITE)  # Clear the screen
    pygame.draw.circle(screen, RED, (ball_x, int(ball_y)), ball_radius)

    pygame.display.flip()  # Update the display

    clock.tick(60)  # Limit the frame rate to 60 FPS

# Quit Pygame
pygame.quit()
sys.exit()
```

```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Bouncing Ball")

# Colors
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# Clock for controlling the frame rate
clock = pygame.time.Clock()

# Ball properties
ball_radius = 20
ball_x = WIDTH // 2
ball_y = ball_radius
ball_velocity_y = 0
ball_velocity_x = 0
ball_speed = 5
jump_strength = -10

gravity = 0.5
bounce_damping = 0.8  # Energy lost on each bounce

# Ground level
ground_y = HEIGHT - ball_radius

running = True
while running:
    # Event handling
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Key states for movement
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        ball_velocity_x = -ball_speed
    elif keys[pygame.K_RIGHT]:
        ball_velocity_x = ball_speed
    else:
        ball_velocity_x = 0

    # Jump when up arrow is pressed
    if keys[pygame.K_UP] and ball_y >= ground_y:
        ball_velocity_y = jump_strength

    # Physics update
    ball_velocity_y += gravity
    ball_y += ball_velocity_y
    ball_x += ball_velocity_x

    # Bounce when hitting the ground
    if ball_y >= ground_y:
        ball_y = ground_y
        ball_velocity_y = -ball_velocity_y * bounce_damping

        # Stop bouncing if velocity is very low
        if abs(ball_velocity_y) < 1:
            ball_velocity_y = 0

    # Prevent ball from moving out of bounds horizontally
    if ball_x - ball_radius < 0:
        ball_x = ball_radius
    if ball_x + ball_radius > WIDTH:
        ball_x = WIDTH - ball_radius

    # Drawing
    screen.fill(WHITE)  # Clear the screen
    pygame.draw.circle(screen, RED, (ball_x, int(ball_y)), ball_radius)

    pygame.display.flip()  # Update the display

    clock.tick(60)  # Limit the frame rate to 60 FPS

# Quit Pygame
pygame.quit()
sys.exit()
```
